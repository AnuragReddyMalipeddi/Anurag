# Finding the Relation Between SAT Scores and Demographics in NYC

- In USA, efficacy of standardized tests is always controversial in the sense that it is biased towards a certain group. We want to find out to what extent it is the case? The goal of this project is to analyse relationship between SAT scores and race,gender,income and other attributes related to students.

- SAT is taken by high school students and hence our focus will be on high schools. NYC is made of five boroughs and many districts. Aggregating data district wise will allow us to look at district by district differences. Each school is identified by district borough number(DBN). Many interrelated datasets need to be cleaned and combined into one dataset. Further various demographic factors like sex, colour, race will be analysed to get a better understanding about their relationship with sat results.

- Upon analysing, we found that natives (white people) tend to have upper hand in the exam which is self explanatory. Surprisingly schools with high total enrollment found positive correlation with sat scores. Even though the reason is not clear, females performed better than the males. Safety of student-teachers is of paramount importance when it comes to good sat scores. 

## Dataset Info:

## Reading the datasets

```python
import pandas as pd
```

```python
#Reading csv files
datafiles = [                  #filenames
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]

#creating an empty dictionary to store the info of all files in the form of keys(filename) and values(dataframe)
#It would be better to access all files in one variable instead of using individual filenames
data={}  #To store sat and demographic data

#reading files into dictionary
for file in datafiles:
    file_path = "schools/{0}".format(file) # All the files are under school folder
    file_info = pd.read_csv(file_path)
    key = file.replace(".csv","")
    data[key] = file_info

```

```python
# Reading survey files which are txt format

all_survey = pd.read_csv("schools/survey_all.txt",delimiter ='\t',encoding="windows-1252") #survey from all schools involving parents,teachers,students
survey_75 = pd.read_csv("schools/survey_d75.txt",delimiter ='\t',encoding="windows-1252") # survey conducted by NYC district 75 which takes of students with significant challenges

print(all_survey.columns)
print(survey_75.columns)

# Combining two surveys. They contain same column headers

survey = pd.concat([all_survey,survey_75],axis=0)
```

# Cleaning the datasets

```python
# survey datasets
#changing the field name dbn to DBN.

survey["DBN"] = survey["dbn"]

# from the data dictionary mentioned above, I found the below columns to be important for our analysis.

survey_fields = [
    "DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_11", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]

survey = survey.loc[:,survey_fields]

print(survey.head())

#Combining survey data with rest of the data.

data["survey"] = survey
```

```python
#hs_directory
# As discussed earlier hs_directory's column dbn needs to be renamed.

data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]
```

```python
# locations coordinates needs to be retrieved from hs_directory which are in "location 1" column.
# Its helps us to map schools and uncover any geographic pattern.

print(data["hs_directory"]["Location 1"].head())    #locations coordinates are placed in the end in the form of \n(40.67,45,98)

import re
def find_lat(loc):                             #getting the latitude
    coord = re.findall(r"\(.+\)",loc)          
    lat = coord[0].split(",")[0].replace("(","")
    return lat

data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(find_lat)

def find_lon(loc):                              #getting the longitude
    coord = re.findall(r"\(.+\)",loc)
    lon = coord[0].split(",")[1].replace(")","")
    return lon

data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(find_lon)

print(data["hs_directory"].head())
```

```python
data["class_size"].head()
print(data["class_size"].info())
```

- class_size does not have dbn 
- But if we look at DBN in SAT_results it is the combination of CSD and Schoolcode but CSD needs to be padded with 0.

```python
def pad_csd(element):
    return str(element).zfill(2)    # CSD is being padded with zeroes

data["class_size"]["pad_csd"] = data["class_size"]["CSD"].apply(pad_csd)

data["class_size"]["DBN"] = data["class_size"]["pad_csd"] + data["class_size"]["SCHOOL CODE"] #School code and padded csd are merged to get DBN

data["class_size"].head()


    
```

```python
# In SAT_results individual marks needs to be changed to numeric type and clubbed to get total SAT score.

cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']

for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c],errors="coerce")
    
data["sat_results"]["sat_score"] = data["sat_results"][cols[0]] + data["sat_results"][cols[1]] + data["sat_results"][cols[2]] 

data["sat_results"].head()
```

```python
# converting AP(Advance Placement) test scores to numeric

data["ap_2010"].info()

# AP Test Takers, Total Exams Taken, Number of Exams with scores 3 4 or 5 need to be converted to numeric

cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

for c in cols:
    data["ap_2010"][c] = pd.to_numeric(data["ap_2010"][c],errors = "coerce")
    
```

# Condense the datasets
- class_size,graduation, demographics have duplicate DBN's which need to be condensed to get unique DBN's for merging the datasets further with sat_results(which has unique DBN)

```python
#CLASS SIZE

data["class_size"].head()
```

- Duplicate DBN's are due to same school name.
- Each school has various program types and grades

```python
data["class_size"]['GRADE '].unique()
```

- Since we are only concerned with high school, we will be only working with 09-12 grade.

```python
data["class_size"]['PROGRAM TYPE'].unique()
data["class_size"]['PROGRAM TYPE'].value_counts()
```

- General education is largest category and hence we will only work with those to trim down the data further.

```python
# Selecting rows where grade is 09-12 and program_type is gen_ed

class_size = data["class_size"]
class_size = class_size.loc[(class_size['GRADE ']=="09-12")&(class_size['PROGRAM TYPE']=='GEN ED')]

class_size.shape
class_size.head()
```

- still the DBN values are non unique.
- Its due to CORE SUBJECT (MS CORE and 9-12 ONLY)	CORE COURSE (MS CORE and 9-12 ONLY)

```python
class_size["CORE SUBJECT (MS CORE and 9-12 ONLY)"].unique()
class_size["CORE COURSE (MS CORE and 9-12 ONLY)"].unique()
```

- core subject includes the course for various class levels 9,10,11 and 12
- we can take the average of these courses to get the overall picture for that subject by using groupby function.

```python
import numpy as np

class_size = class_size.groupby("DBN").agg(np.mean) #using groupby to eliminate duplicates 
data["class_size"] = class_size

data["class_size"].head()
```

```python
# DEMOGRAPHICS

data["demographics"].head()
```

- Duplicacy of demographics dataset is due to school_year column.
- we only want to work with 20112012 data so that it matches with SAT results data(it belongs to 2012)

```python
data["demographics"] = data["demographics"].loc[data["demographics"]["schoolyear"]==20112012]
data["demographics"].head()
data["demographics"].shape
```

```python
# GRADUATION

data["graduation"].head()
data["graduation"]["Demographic"].unique()
data["graduation"]["Cohort"].unique()
```

- DBN in graduation is non-unique owing to Demographics and cohort
-  Cohort seems to be the year in which data is taken. We will keep the latest year which is 2006.
- Demographic seems to be the specific demographic area. We will go with total cohort


```python
data["graduation"] = data["graduation"].loc[(data["graduation"]["Demographic"]=="Total Cohort")&(data["graduation"]["Cohort"]=="2006")]

data["graduation"].head()
data["graduation"].shape
```

# Merging the Datasets

- We will use DBN to identify matching rows across datasets to combine data
- There might be many issues with this method like DBN in one dataset might not be in other because of the data being from different years.
-  we will merge two datasets at a time and use various merge strategies in each iteration.
- Because this project is concerned with relation between demographic data and SAT scores, we want to preserve all rows from SAT_results while minimizing null values.



```python
# Both ap_2010 and graduation has many missing DBN values since they are from different years(2010) when compared with SAT_results(2012) and hence we use left join with SAT_results
combined = data["sat_results"]
combined.head(5)
combined.shape
combined = combined.merge(data["ap_2010"],on="DBN",how="left")
combined = combined.merge(data["graduation"],on="DBN",how="left")
combined.head(5)
combined.shape
```

```python
# class_size, hs_directory,demographics,survey will be merged using inner join since the year of data is in sync with SAT_results and also the most valuable data for our analysis.

merge_sets = ["class_size","demographics","survey", "hs_directory"]

for dataset in merge_sets:
    combined = combined.merge(data[dataset],on="DBN",how="inner")
    combined.head(5)
    combined.shape
```

```python
combined.head(5)
combined.shape
```

# Filling the Null Values

```python
pd.options.display.max_rows = 4000 
combined.isnull().sum()
```

- we will replace the null values with their respective columns mean values.
- leftover values would be replcaed by 0

```python
mean = combined.mean()
combined = combined.fillna(mean) #null values are replaced by their respective mean values
combined = combined.fillna(0)
```

```python
# As we want to base our analysis on districts, we want to add a district column to the combined dataset
# district is the first two characters in the DBN

combined["school_dist"] = combined["DBN"].str[0:2]
```
