```python
reviews_max = {}

for row in android:
    app = row[0]
    n_reviews = float(row[3])
    if app in reviews_max and reviews_max[app] < n_reviews:
        reviews_max[app] = n_reviews
    if app not in reviews_max:
        reviews_max[app] = n_reviews
        
print(len(reviews_max))
```

```python
android_clean = []
already_added = []

for row in android:
    app = row[0]
    n_reviews = float(row[3])
    if n_reviews == reviews_max[app] and app not in already_added:
        android_clean.append(row)
        already_added.append(app)
        
print(len(android_clean))
```

- reviews_max dictionary is created to store apps with highest reviews
- android_clean list of lists is created by selecting apps with highest reviews (using reviews_max dictionary)

```python
def check_if_english_name(app_name):
    count = 0
    for character in app_name:
        if ord(character)>127:
            count += 1
    if count>3:
        return False
    else:
        return True

print(check_if_english_name('Instagram'))
print(check_if_english_name('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(check_if_english_name('Docs To Go™ Free Office Suite'))
print(check_if_english_name('Instachat 😜'))
```

```python
android_english =[]
ios_english = []

for row in android_clean:
    if check_if_english_name(row[0]):
        android_english.append(row)
        
for row in ios:
    if check_if_english_name(row[1]):
        ios_english.append(row)
        
        
print(len(android_english))
print(len(ios_english))
```

```python

def freq_table(dataset,index):
    freq_dict = {}
    freq_dict_per = {}
    total = 0
    for row in dataset:
        total += 1
        key = row[index]
        if key in freq_dict:
            freq_dict[key] += 1
        else:
            freq_dict[key] = 1

    for key in freq_dict:
        freq_dict_per[key] = (freq_dict[key]/total)*100
    return freq_dict_per
```

```python
def display_table(dataset,index):
    table = freq_table(dataset,index)
    table_display = []
    for key in table:
        key_value_as_tuple = (table[key],key)
        table_display.append(key_value_as_tuple)
    sorted_table_display = sorted(table_display,reverse=True)

    for element in sorted_table_display:
        print(element[1],":",element[0])
```

```python
prime_genre_freq_table = freq_table(ios_final,-5)
prime_genre_freq_table
```

```python
avg_tot_ratings_per_genre = {}
for key in prime_genre_freq_table:
    total_num_ratings_per_genre = 0
    apps_per_genre = 0
    for row in ios_final:
        if row[-5] == key:
            total_num_ratings_per_genre += float(row[5])
            apps_per_genre += 1
    avg_tot_ratings_per_genre[key] = total_num_ratings_per_genre/apps_per_genre

def display_users(dictionary):   
    table_display = []
    for key in dictionary:
        key_value_as_tuple = (dictionary[key],key)
        table_display.append(key_value_as_tuple)
    sorted_table_display = sorted(table_display,reverse=True)

    for element in sorted_table_display:
        print(element[1],":",element[0])
            
display_users(avg_tot_ratings_per_genre)
```

# Lets explore Google Play Store users using installs column


- Since this column contains "+" and "," characters, we need to replace them with spaces and perform numerical operations after converting them to float type.

```python
average_installs_per_category = {}
for key in category_freq_table:
    total_installs = 0
    apps_per_category=0
    for row in android_final:
        if row[1] == key:
            installs = row[5]
            installs = installs.replace("+","")
            installs = installs.replace(",","")
            installs = float(installs)
            total_installs += installs
            apps_per_category += 1
    average_installs_per_category[key] = total_installs/apps_per_category
    
display_users(average_installs_per_category)
```
