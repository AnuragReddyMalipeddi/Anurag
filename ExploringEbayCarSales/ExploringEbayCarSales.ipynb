# Cleaning the Dataset

# Cleaning the column names

```python
old_column_names = autos.columns 
old_column_names
```

```python

column_map = {
    'yearOfRegistration' : 'registration_year',
    'monthOfRegistration':'registration_month',
    'notRepairedDamage'  :'unrepaired_damage',
    'dateCreated'        : 'ad_created',
    'dateCrawled'        : 'date_crawled',
    'offerType'          : 'offer_type',
    'vehicleType'        : 'vehicle_type',
    'fuelType'           : 'fuel_type',
    'postalCode'         : 'postal_code',
    'nrOfPictures'       : 'nr_of_pictures',
    'lastSeen'           : 'last_seen'
}

# renaming the columns

autos.rename(column_map,inplace=True,axis=1)
```

```python
new_column_names = autos.columns
new_column_names
```

- Few column names were not depicting the actual description according to data dictionary and hence they were renamed
- Also the column names with camel case were changed to snake case.

```python
autos.describe(include="all")
```

- Price and odometer are in string type which need to be converted to int.
- nr_of_pictures column does not have any other value other than '0' and hence it needs to be dropped.

# Dropping nr_of_pictures column

```python
autos["nr_of_pictures"].unique()
```

```python
autos.drop("nr_of_pictures",axis=1)
```

# Converting datatypes of price and odometer

```python
autos["price"] = autos["price"].str.replace("$","").str.replace(",","").astype(float)
    
```

```python
autos["odometer"] = autos["odometer"].str.replace("km","").str.replace(",","").astype(float)
```

```python
autos.rename({"odometer":"odometer_km","price":"price$"},inplace=True,axis=1) #renmaing odometer and price columns
```

```python
autos.head()
```

# Analysing price and odometer_km columns

- We want to see for anomalies in these fileds to filter out unreasonable rows

# Price

```python
print(autos["price$"].unique().shape)
print(autos["price$"].describe())
autos["price$"].value_counts().head(20)
```

```python
autos["price$"].value_counts().sort_index(ascending=False).head(20) # top 20 price list
```

```python
autos["price$"].value_counts().sort_index(ascending=True).head(20) #Bottom 20 price list
```

-  we find that price column has highest value of 99999999.0 and lowest of 0
- Considering ebay website where bids start at 1, 0 is appropriate.
- Also anything above 350,000 for a used car is not accuarate and hence need to be dropped.

- Excluding rows with price greater than 351,000

```python
autos["price$"] = autos[autos["price$"].between(1,351000)]
```

# Exploring Brand Data


```python
autos["brand"].value_counts(normalize=True)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True)
top_brands = brand_counts[brand_counts>0.05].index
top_brands
```

```python
brand_mean_prices = {}

for brand in top_brands:
    brand_only = autos[autos["brand"]==brand]
    mean_price = brand_only["price$"].mean()
    brand_mean_prices["brand"] = mean_price
```
